```python
# Task 1: Superposition

# Task 1: Superposition
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

# Create a 1-qubit circuit
qc1 = QuantumCircuit(1, 1)
qc1.h(0)           # Apply Hadamard gate
qc1.measure(0, 0)  # Measure

# Use AerSimulator (new syntax)
simulator = AerSimulator()
result1 = simulator.run(qc1, shots=1024).result()
counts1 = result1.get_counts(qc1)

print("Superposition Results:", counts1)
print("Expected: Approximately equal counts for '0' and '1'")
```

```python
# Task 2: Entanglement

from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

# Create 2-qubit circuit
qc2 = QuantumCircuit(2, 2)
qc2.h(0)          # Hadamard gate on qubit 0
qc2.cx(0, 1)      # CNOT → entanglement
qc2.measure([0, 1], [0, 1])

# Simulate
simulator = AerSimulator()
result2 = simulator.run(qc2, shots=1024).result()
counts2 = result2.get_counts(qc2)

# Text output only
print("Entanglement Results:", counts2)
print("Expected: Correlated outcomes — mostly '00' and '11'")
```
